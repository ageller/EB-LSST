# Exploring the 2018a OpSim database

```python
import sqlite3
import numpy as np
```

```python
db = sqlite3.connect("baseline2018a.db")
```

```python
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

### ObsHistory is the replacement for minion's summary 

```python
cursor.execute("PRAGMA table_info(ObsHistory)")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT Field_fieldid, observationStartTime, filter, fiveSigmaDepth FROM ObsHistory") 
summaryCursor = np.array(cursor.fetchall()) #NOTE: this takes a LONG time
```

```python
print(len(summaryCursor[:,1].astype('float')))
```
