# Create the open cluster file for latex, to include in the paper

I compiled the open cluster catalog here: https://github.com/ageller/compileOCs , 

I think I will include all the columns in the clean file


Note: masses assume a mean stellar mass of 0.5 Msun (for literature sources that provide number of stars)

Note: rhm is a bit tricky.  Maybe I should just call this a radius estimate?  If the rc is provided, I convert this to rhm using a Plummer model, others provide a diameter measurement, which I divide by two.

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)

from astropy import units, constants
from astropy.coordinates import SkyCoord

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
OC_df = pd.read_csv('OCcompiled_clean.csv')
OC_df
```

```python
OC_df.rename(columns = {'RA':'RA[hr]','Dec':'Dec[deg]','rhm[pc]':'r[pc]','err_rhm[pc]':'err_r[pc]','N_rhm[pc]':'N_r[pc]'}, inplace=True)
OC_df.replace('_', ' ', regex=True, inplace=True)
OC_df.round(1, inplace_True)
OC_df
```

```python
OC_df.to_latex('OCtable.tex',na_rep='nodata', index=False)
```
