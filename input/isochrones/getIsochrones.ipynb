# Get isochrones for each cluster

Using ezpadova : https://github.com/mfouesneau/ezpadova

```python
import pandas as pd
import numpy as np
import time


from ezpadova import parsec
```

```python
GCdf = pd.read_csv('/Users/ageller/WORK/LSST/onGitHub/EBLSST/input/clusters/GCdataForEBLSST.csv')
GCdf
```

```python
#input is age (yr), metalicity
#[Fe/H] = log(Zstar/Zsun) = log(Zstar) - log(Zsun)
#log(Zstar) = [Fe/H] + log(Zsun)
#isochrones define for 0.0001â‰¤Zâ‰¤0.06 
logZsun = np.log10(0.0152)
```

```python
for index, row in GCdf.iterrows():
    if (index > 43):
        Z = np.clip(10.**(row['[Fe/H]'] + logZsun), 0.0001, 0.06)
        age = np.clip(row['age[Myr]']*1e6,0,13.5e9)
        ID = row['ID'].replace(' ','_')
        print(index, ID, Z, age)

        r = parsec.get_one_isochrone(age, Z, phot='lsst')
        r.write('/Users/ageller/WORK/LSST/onGitHub/EBLSST/input/isochrones/GCs/'+ID+'.csv')
#         if (index > 0):
#             break
        print('sleeping...')
        time.sleep(10)
```
